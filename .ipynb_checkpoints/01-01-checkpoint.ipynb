## 1.2　散布図を描く
http://kogolab.chillout.jp/elearn/icecream/chap1/sec2.html

```r
# 折れ線グラフ
plot1 <- ggplot(data=df, mapping=aes(x=日付, y=客数)) +
    xlim(1, 14) + ylim(0, 700) +
    geom_point(size=4) + geom_line() +
    labs(title="A店における客数の変化") +
    theme(
        plot.title=element_text(size=20, hjust=0.5),
        axis.title.x=element_text(size = 16),
        axis.text.x=element_text(size=12),
        axis.title.y=element_text(size = 16),
        axis.text.y=element_text(size=12),
    )
plot1
```

```r
# 散布図
plot2 <- ggplot(data=df, mapping=aes(x=最高気温, y=客数)) +
    xlim(25, 35) + ylim(200, 700) +
    geom_point(size=4) +
    labs(title="A店における気温と客数の関係") +
    theme(
        plot.title=element_text(size=20, hjust=0.5),
        axis.title.x=element_text(size = 16),
        axis.text.x=element_text(size=12),
        axis.title.y=element_text(size = 16),
        axis.text.y=element_text(size=12),
    )
    
plot2
```

最高気温が高くなると客数も増えている。  
また、データが週末と平日の二つのグループに分かれていて、週末は客数が多くなっている。

## 1.3　散布図のパターン
* xの値が大きくなればyの値が大きくなるようなデータは「正の相関」があるといい、散布図の点は右上がりになる。
* xの値が大きくなればyの値が小さくなるようなデータは「負の相関」があるといい、散布図の点は右下がりになる。
* xの値とyの値に関係がないデータは相関がない（無相関）といい、散布図の点はバラバラに散らばる。
